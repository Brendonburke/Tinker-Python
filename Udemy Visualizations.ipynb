```python
df2=pd.DataFrame(np.random.randint(100, size=(100,2)),columns=['male', 'female'])
df2.head()
```

```python
df2['male'].plot.hist()
```

```python
df2['female'].plot.hist()
```

```python
df2.plot.hist(alpha = .5)
```

```python
df2.plot.hist(stacked=True)
```

```python
df2.plot.box()
```

```python
df2.plot.scatter('male', 'female')
```

```python
#create a new column to weight the points
df2['score']=df2.male*.3+df2.female*.7
df2.head()
```

```python
df2.plot.scatter(x='male', y='female', s=df2.score)
```
